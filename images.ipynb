```python
%pip install scikit-image matplotlib
# Import necessary libraries
import numpy as np
from skimage.feature import corner_harris, corner_peaks
from skimage.color import rgb2gray
import matplotlib.pyplot as plt
from skimage.io import imread

# Define the corner visualization function (fixed from earlier)
def show_corners(corners, image):
    fig = plt.figure()
    plt.gray()
    plt.imshow(image)
    y_corners, x_corners = zip(*corners)
    plt.plot(x_corners, y_corners, 'or')  # 'or' = red circles
    plt.xlim(0, image.shape[1])
    plt.ylim(image.shape[0], 0)
    fig.set_size_inches(np.array(fig.get_size_inches()) * 1.5)
    plt.show()

# Step 1: Load an image (replace with your image path)
image_path = 'lion.jpg'  # Example: 'chessboard.jpg'
img_color = imread(image_path)  # Load in RGB format

# Step 2: Convert to grayscale (Harris works on single-channel images)
img_gray = rgb2gray(img_color)

# Step 3: Detect corners using Harris corner detection
corners = corner_peaks(
    corner_harris(img_gray),
    min_distance=5,  # Minimum distance between corners
    threshold_rel=0.1  # Relative threshold for corner response
)

# Step 4: Visualize the detected corners  
show_corners(corners, img_color)
```
